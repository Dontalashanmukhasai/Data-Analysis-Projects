```python
results['State'] = results['State'].str.strip().str.lower()
locations['Location name'] = locations['Location name'].str.strip().str.lower()
merged_data = pd.merge(results,locations,left_on = 'State',right_on = 'Location name',how ='left')
merged_data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
state_ad_spend = merged_data.groupby('State')['Amount spent (INR)'].sum().reset_index()
state_ad_spend = state_ad_spend.sort_values(by='Amount spent (INR)', ascending=False).reset_index(drop=True)
state_ad_spend
```

```python
import numpy as np
df = pd.DataFrame(state_ad_spend['State'])
plt.figure(figsize=(12,6))
fig = plt.bar(state_ad_spend['State'],state_ad_spend['Amount spent (INR)'],color='skyblue', edgecolor='blue',width=0.8)
plt.xticks(rotation=90)
plt.title('Total Ad Spend By Each State')
plt.xlabel('State')
plt.ylabel('Ad Spend(INR)')
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('advertisers.csv')
df.head()
```

```python
df['Amount spent (INR)'] = pd.to_numeric(df['Amount spent (INR)'], errors='coerce')
df.dropna(subset=['Amount spent (INR)'], inplace=True)
party_ad_spend = df.groupby('Page name')['Amount spent (INR)'].sum().sort_values(ascending=False)
top5_parties = party_ad_spend.head(5).reset_index()
top5_parties
```

```python
labels = top5_parties['Page name']
values = top5_parties['Amount spent (INR)']
plt.figure(figsize=(10, 7))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Amount Spent by Page Name')
plt.show()
```

```python
results = pd.read_csv('results.csv')
results.head()
```

```python
results['State'] = results['State'].str.strip().str.lower()
locations['Location name'] = locations['Location name'].str.strip().str.lower()
merged_constituency_data = results.merge(
    locations,
    left_on='State',
    right_on='Location name',
    how='left'
)
```

```python
filtered_data = merged_constituency_data.dropna(subset=['Amount spent (INR)', 'Polled (%)', 'State'])
filtered_data
```

```python
fig = plt.scatter(filtered_data['Amount spent (INR)'], 
                     filtered_data['Polled (%)'],alpha=0.7)
plt.title('Ad Spend and Voter Turnout by Parliamentary Constituency')
plt.xlabel('Ad spend (INR)')
plt.ylabel('Voter Turnout(%)')
plt.show()
```

```python
fig = plt.subplots(figsize=(10, 7))
n, bins, patches = plt.hist(merged_data['Amount spent (INR)'], bins=30, edgecolor='black')
plt.boxplot(x=merged_data['Amount spent (INR)'])
plt.xlabel('Ad Spend (INR)')
plt.ylabel('Count')
plt.title('Distribution of Ad Spend(INR)')
plt.tight_layout()
plt.show()
```

```python
phase_analysis = merged_data.groupby('Phase').agg({
    'Amount spent (INR)': 'sum',
    'Polled (%)': 'mean'}).reset_index()
phase_analysis
```

```python
fig = plt.subplots(figsize=(10, 6))
bars=plt.bar(phase_analysis['Phase'], phase_analysis['Amount spent (INR)'], color='indianred', label='Ad Spend (INR)')
plt.xlabel('Election Phase')
plt.ylabel('Ad Spend (INR)', color='indianred')
plt.tick_params(axis='y', labelcolor='indianred')

plt1 = plt.twinx()
line = plt1.plot(phase_analysis['Phase'], phase_analysis['Polled (%)'], color='lightsalmon', marker='o', label='Voter Turnout (%)')
plt1.set_ylabel('Voter Turnout (%)', color='lightsalmon')
plt1.tick_params(axis='y', labelcolor='lightsalmon')
plt.title('Ad Spend and Voter Turnout by Election Phase')
plt.tight_layout()
plt.show()
```
